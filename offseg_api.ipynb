```python
#管理者ログイン(管理者のみ)
import json

def update_id_token(id_token,user_id,company_id,refresh_token):
    REFRESH_TOKEN_URL = "https://exsyscoopapi.offseg.obvious.jp/api/public/login/reissue-token"
    request_headers = {'id_token': id_token ,'operation_user_id': str(user_id),'operation_company_id': str(company_id),'Content-Type': 'application/json'}
    request_body = {'refresh_token': refresh_token}

    # POSTリクエストを、リクエストボディ付きで送信する
    response = requests.put(REFRESH_TOKEN_URL, headers=request_headers,data=json.dumps(request_body))

    # レスポンスボディを出力する
    pprint.pprint(response.json())
 
    new_id_token = response.json()['content']['id_token']
    
    return new_id_token
```

```python
## アドミン情報を取得
## 実行ユーザIDや実行カンパニーID、トークン情報を取得
import requests
import pprint
import json
import os

company_code = os.environ['OFFSEG_COMPANY_CODE']
user_code = os.environ['OFFSEG_USER_CODE']
password = os.environ['OFFSEG_PASSWORD']


def get_admin_info(company_code,user_code,password):
    ADMIN_URL = f"{URL}/admin/login"
    request_body = {'company_code': company_code, 'user_code': user_code, 'password': password}

    # POSTリクエストを、リクエストボディ付きで送信する
    response = requests.post(ADMIN_URL, json=request_body)

    id_token = response.json()['content']['authentication_result']['token_info']['id_token']
    refresh_token = response.json()['content']['authentication_result']['token_info']['refresh_token']
    company_id = response.json()['content']['search_result']['company_id']
    user_id = response.json()['content']['search_result']['user_id']

    # レスポンスボディを出力する
    #pprint.pprint(response.json())

    return id_token,refresh_token,company_id,user_id
```

```python
#定常センサデータ取得

def get_ticket_info_const(car_id_list,id_token,user_id,company_id,search_periodic_from,search_periodic_to):
    SENSOR_POST_URL = f"{URL}/periodic_data/acquisition_request"

    request_headers = {'id_token': id_token ,
                    'operation_user_id': str(user_id),
                    'operation_company_id': str(company_id)
                    }
    request_body = {
                    'request_id':1,
                    'search_periodic_from':search_periodic_from,
                    'search_periodic_to':search_periodic_to,
                    'car_id_list':car_id_list,
                    'output_format':'02'#JSON
                    }

    # POSTリクエストを、リクエストボディ付きで送信する
    response = requests.post(SENSOR_POST_URL,headers=request_headers,json=request_body)
    resultcode = response.json()['resultcode']
    # レスポンスボディを出力する
    #pprint.pprint(response.json())
    resultcode = response.json()['resultcode']
    if resultcode==0:
        ticket_id = response.json()['content']['ticket_id']
    else:
        ticket_id = response.json()['content']['messages']
    return resultcode,ticket_id
```

```python
#定常センサデータ取得

def get_ticket_info_trip(car_id_list,id_token,user_id,company_id,search_periodic_from,search_periodic_to):
    TRIP_POST_URL = f"{URL}/periodic_data/trip_acquisition_request"

    request_headers = {'id_token': id_token ,'operation_user_id': str(user_id),'operation_company_id': str(company_id)}
    request_body = {'search_periodic_from':search_periodic_from,
                    'search_periodic_to':search_periodic_to,
                    'request_id':1,
                    'car_id_list':car_id_list,
                    'output_format':'02' #JSON
                    }
    # POSTリクエストを、リクエストボディ付きで送信する
    response = requests.post(TRIP_POST_URL,headers=request_headers,json=request_body)

    # レスポンスボディを出力する
    #pprint.pprint(response.json())
    resultcode = response.json()['resultcode']
    if resultcode==0:
        ticket_id = response.json()['content']['ticket_id']
    else:
        ticket_id = response.json()['content']['messages']
    return resultcode,ticket_id
```

```python
def get_file_info(id_token,user_id,company_id,ticket_id):

#定常センサデータ取得状況確認

    ACQUITION_URL = "https://exsyscoopapi.offseg.obvious.jp/api/public/periodic_data/acquisition_status"

    request_headers = {
        'id_token': id_token ,
        'operation_user_id': str(user_id),
        'operation_company_id': str(company_id)
        }
    request_body = {'request_id':1,
                    'ticket_id':ticket_id
                    }
    # POSTリクエストを、リクエストボディ付きで送信する
    response = requests.post(ACQUITION_URL,headers=request_headers,json=request_body)

    # レスポンスボディを出力する
    #pprint.pprint(response.json())
    resultcode = response.json()['resultcode']
    if resultcode==0:
        file_info = response.json()['content']['file_info']
    else:
        file_info = response.json()['content']['messages']
    return resultcode,file_info
```

```python
# 定常センサデータ収集情報取得
def get_sensor_data_url(id_token,user_id,company_id,ticket_id,file_id):

    FILE_GET_URL = f"{URL}/periodic_data/file_url"

    request_headers = {'id_token': id_token ,'operation_user_id': str(user_id),'operation_company_id': str(company_id)}
    request_body = {'request_id':1,
                    'ticket_id':ticket_id,
                    'file_id':file_id
                    }
    # POSTリクエストを、リクエストボディ付きで送信する
    response = requests.post(FILE_GET_URL,headers=request_headers,json=request_body)

    # レスポンスボディを出力する
    resultcode = response.json()['resultcode']
    if resultcode==0:
        file_url = response.json()['content']['presigned_url']
    else:
        file_url = response.json()['content']['messages']    

    return resultcode,file_url
```

```python
from glob import glob
import json
import pandas as pd
from io import StringIO


def upload_json_file():
    folder_list = glob("./sensor_data/*")
    print(folder_list)
    
    
    for folder in folder_list:
        folder_name = folder.split('\\')[-1]
        print(folder_name)
        json_list = glob(f"{folder}/*.json")
        print(json_list)
        json_text = ""
        for r,json_file in enumerate(json_list):
            with open(json_file,'r',encoding='utf-8') as f:
                json_text += f.read()
        with open("tmp_json_file.json",'w',encoding='utf-8') as f1:
            f1.write(json_text)
        # StringIOオブジェクトでラップ
        json_data = StringIO(json_text)
        # JSONデータを読み込む
        df = pd.read_json(json_data)
        if not os.path.exists('./sensor_csv_data'):
            os.mkdir('./sensor_csv_data')
        df.to_csv(f'./sensor_csv_data/{folder_name}.csv',index=False)
```

```python
import urllib.request
import sys
import time
import tarfile
import pandas as pd


def upload_offseg_sensor_data(date_str,car_id_list,id_token,user_id,company_id):
    #car_id_list = [4051]
    #date_str = '2024091712'#対象の日付を指定
    #1日単位で設定しようとしたが、ファイルサイズが大きくなりすぎのため、1時間単位に修正
    search_periodic_from = f'{date_str}0000'
    search_periodic_to = f'{date_str}5959'


    print('ticket_id ===================================================================================================')
    resultcode_ticket_info,ticket_id = get_ticket_info_const(car_id_list=car_id_list,id_token=id_token,user_id=user_id,company_id=company_id,search_periodic_from=search_periodic_from,search_periodic_to=search_periodic_to)

    if resultcode_ticket_info == 0:
        time.sleep(1)#SLEEPをいれないと、データが取得できない(たぶん生成するまでのラグがある？)
        print('ticket_id:',ticket_id)
        print('file_info ===================================================================================================')
        result_code_file_info,file_infos = get_file_info(id_token=id_token,user_id=user_id,company_id=company_id,ticket_id=ticket_id)
        
        #定常センサデータ取得状況確認
        if result_code_file_info == 0:
            print(file_infos)
            print('file_url ===================================================================================================')
            for file_info in file_infos:
                car_id = file_info['car_id']
                file_id = file_info['file_id']
                resultcode_url,file_url = get_sensor_data_url(id_token=id_token,user_id=user_id,company_id=company_id,ticket_id=ticket_id,file_id=file_id)
                #fileのダウンロード
                if resultcode_url == 0:
                    print(file_url)
                    if not os.path.exists('./tar_file'):
                        os.mkdir('./tar_file')
                    download_file = f'./tar_file/car_id_{car_id}-day_{date_str}.tar.gz'
                    urllib.request.urlretrieve(file_url,"{0}".format(f"{download_file}"))
                    #ダウンロードしたファイルを解凍する
                    json_list = []
                    json_text = ""
                    with tarfile.open(f'{download_file}') as tar:
                        for i in tar.getmembers():
                            if i.name[0] == '/' or i.name[0:2] == '..':
                                exit()
                            json_list.append(i.name)
                        if not os.path.exists(f'./sensor_data'):
                            os.mkdir(f'./sensor_data') 
                        if not os.path.exists(f'./sensor_data/car_id_{car_id}-date_{search_periodic_from}'):
                                os.mkdir(f'./sensor_data/car_id_{car_id}-date_{search_periodic_from}')
                        tar.extractall(path=f'./sensor_data/car_id_{car_id}-date_{search_periodic_from}')
                    
                    #jsonファイルを結合してcsvに変換する
                    for r,json_file in enumerate(json_list):
                        json_file_path = f'./sensor_data/car_id_{car_id}-date_{search_periodic_from}/{json_file}'
                        with open(json_file_path,'r',encoding='utf-8') as f:
                            json_text += f.read()
                    # StringIOオブジェクトでラップ
                    json_data = StringIO(json_text)
                    # JSONデータを読み込む
                    df = pd.read_json(json_data)
                    
                    if not os.path.exists('./sensor_csv_data'):
                        os.mkdir('./sensor_csv_data')
                    df.to_csv(f'./sensor_csv_data/car_id_{car_id}-date_{search_periodic_from}.csv',index=False)

                else:
                    print(file_url)
        else:
            print(f"{file_infos}")
    else:
        print(f"{ticket_id}")
```
